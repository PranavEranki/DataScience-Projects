```python
# imports
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("imgs/github_logo.jpg")
```

```python
grayscale = cv2.imread("imgs/github_logo.jpg", cv2.IMREAD_GRAYSCALE) # Simpler, used for computer vision
```

Popular options:


IMREAD_ ...

-COLOR

-UNCHANGED

-GRAYSCALE

```python
# showing the image
plt.imshow(img)
plt.title("Github logo pic")
plt.show()
```

```python
plt.imshow(grayscale, cmap = 'gray', interpolation = 'bicubic')
plt.plot([50,100,200,300, 30], [40,100,4,342, 324], 'c')
plt.show()
```

```python
cv2.imwrite("imgs/gray_github_logo.JPG", grayscale)
```
